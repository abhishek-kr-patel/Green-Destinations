```python
import pandas as pd
```

```python
df=pd.read_csv('greendestination (1).csv')
```

```python
df.head(10)
```

```python
df[['Age', 'YearsAtCompany', 'MonthlyIncome']].describe()
```

```python
# Check the column names to identify the one related to attrition
print(df.columns)
```

```python
# Calculate the attrition rate
total_employees = df.shape[0]  
attrition_count = df[df['Attrition'] == 'Yes'].shape[0]  
attrition_rate = (attrition_count / total_employees) * 100
print('Total Employees:',total_employees)
print('How much employees who left:', attrition_count)
print('Attrition Rate:',round(attrition_rate),'%')
```

```python
# Calculate the average age of employees who left vs. those who stayed
age_attrition = df.groupby('Attrition')['Age'].mean()

print("Average Age of Employees:")
print(age_attrition)
```

```python
# Calculate the average years at the company for employees who left vs. those who stayed
years_at_company_attrition = df.groupby('Attrition')['YearsAtCompany'].mean()

print("Average Years at Company of Employees:")
print(years_at_company_attrition)
```

```python
# Calculate the average income for employees who left vs. those who stayed
income_attrition = df.groupby('Attrition')['MonthlyIncome'].mean()

print("Average Income of Employees:")
print(income_attrition)
```
